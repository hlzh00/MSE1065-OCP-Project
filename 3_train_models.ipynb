```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


h_X_train, h_X_test, h_y_train, h_y_test = train_test_split(
    h_X, 
    h_y, 
    test_size=0.2, 
    random_state=1065)

scaler = StandardScaler()
h_X_train_scaled = scaler.fit_transform(h_X_train)
h_X_test_scaled = scaler.transform(h_X_test)

X = h_X_train_scaled
y = h_y_train.to_numpy().ravel()

# no_ads_X_train, no_ads_X_test, no_ads_y_train, no_ads_y_test = train_test_split(
#     encoded_no_ads_X, 
#     no_ads_y, 
#     test_size=0.2, 
#     random_state=1065)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import cross_val_score, learning_curve
from sklearn.metrics import r2_score

# 1. Check if it's a data quantity issue - Learning Curves
train_sizes, train_scores, val_scores = learning_curve(
    Ridge(alpha=1.0), X, y, 
    cv=5, 
    train_sizes=np.linspace(0.1, 1.0, 10),
    scoring='r2',
    n_jobs=-1
)

plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_scores.mean(axis=1), label='Training', linewidth=2)
plt.plot(train_sizes, val_scores.mean(axis=1), label='Validation', linewidth=2)
plt.fill_between(train_sizes, 
                 val_scores.mean(axis=1) - val_scores.std(axis=1),
                 val_scores.mean(axis=1) + val_scores.std(axis=1),
                 alpha=0.2)
plt.xlabel('Training Set Size', fontsize=12)
plt.ylabel('R Score', fontsize=12)
plt.title('Learning Curve - Is More Data Needed?', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# If validation curve has plateaued: more data won't help much
# If curves are converging but both low: you've hit the ceiling
# If curves still diverging: more data might help


# 2. Check for non-linear relationships in residuals
lr = LinearRegression()
lr.fit(X, y)
y_pred = lr.predict(X)
residuals = y - y_pred

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Residual plot
axes[0].scatter(y_pred, residuals, alpha=0.5, s=20)
axes[0].axhline(y=0, color='r', linestyle='--', linewidth=2)
axes[0].set_xlabel('Predicted Values', fontsize=12)
axes[0].set_ylabel('Residuals', fontsize=12)
axes[0].set_title('Residual Plot - Random = Good', fontsize=14)
axes[0].grid(True, alpha=0.3)

# Q-Q plot for normality
from scipy import stats
stats.probplot(residuals, dist="norm", plot=axes[1])
axes[1].set_title('Q-Q Plot - Check Assumptions', fontsize=14)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# If residuals show clear patterns: non-linearity exists but models can't capture it
# If residuals are random: linear model is appropriate


# 3. Feature importance - are features actually predictive?
from sklearn.linear_model import LassoCV

lasso = LassoCV(cv=5, random_state=42, max_iter=10000)
lasso.fit(X, y)

# Count non-zero coefficients
n_selected = np.sum(np.abs(lasso.coef_) > 1e-5)
print(f"\nLasso selected {n_selected}/{X.shape[1]} features")
print(f"Lasso CV R: {lasso.score(X, y):.3f}")

# Plot coefficient magnitudes
coef_sorted = np.sort(np.abs(lasso.coef_))[::-1]
plt.figure(figsize=(12, 5))
plt.bar(range(len(coef_sorted)), coef_sorted)
plt.xlabel('Feature Rank', fontsize=12)
plt.ylabel('|Coefficient|', fontsize=12)
plt.title('Lasso Feature Importance - How Many Features Matter?', fontsize=14)
plt.grid(True, alpha=0.3)
plt.show()

# If only a few features have large coefficients: most features are noise


# 4. Correlation with target
correlations = np.array([np.corrcoef(X[:, i], y)[0, 1] for i in range(X.shape[1])])
correlations_abs = np.abs(correlations)

print(f"\nFeature-target correlations:")
print(f"  Max: {correlations_abs.max():.3f}")
print(f"  Median: {np.median(correlations_abs):.3f}")
print(f"  Features with |corr| > 0.3: {np.sum(correlations_abs > 0.3)}")
print(f"  Features with |corr| > 0.2: {np.sum(correlations_abs > 0.2)}")

plt.figure(figsize=(12, 5))
plt.hist(correlations_abs, bins=50, edgecolor='black')
plt.xlabel('|Correlation with Target|', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Distribution of Feature-Target Correlations', fontsize=14)
plt.axvline(x=0.3, color='r', linestyle='--', label='|r| = 0.3')
plt.axvline(x=0.2, color='orange', linestyle='--', label='|r| = 0.2')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# If most correlations < 0.2: features are weakly predictive
```

```python
from sklearn.linear_model import Lasso, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_validate

rf = RandomForestRegressor(
    n_estimators=100,        # Enough trees
    max_depth=5,             # KEY: Severely limit depth (try 3-7)
    min_samples_split=50,    # Require many samples to split (try 30-100)
    min_samples_leaf=20,     # Require many samples per leaf (try 15-50)
    max_features='sqrt',     # Only use sqrt(93)10 features per split
    random_state=42
)

# regression models (not classifiers)
models = {
    "lasso": Lasso(alpha=0.1, max_iter=100000),
    "ridge": Ridge(alpha=1.0),
    "rf": rf,
}

# choose a regression model to evaluate (change as needed)
model = models["rf"]

# use regression scorers; cross_validate accepts multiple scorers
cv_results = cross_validate(
    model,
    h_X_train_scaled,
    h_y_train.to_numpy().ravel(),
    cv=5,
    return_train_score=True,
    scoring=('r2', 'neg_mean_absolute_error')
)

train_r2 = cv_results['train_r2'].mean()
val_r2 = cv_results['test_r2'].mean()
train_mae = -cv_results['train_neg_mean_absolute_error'].mean()
val_mae = -cv_results['test_neg_mean_absolute_error'].mean()

print(f"Train R: {train_r2:.3f}, Validation R: {val_r2:.3f}")
print(f"Train MAE: {train_mae:.3f}, Validation MAE: {val_mae:.3f}")
print(f"R Gap: {train_r2 - val_r2:.3f}")
```
